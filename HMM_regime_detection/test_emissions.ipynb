```python
import numpy as np
import hmm
import emission_model
import matplotlib.pyplot as plt
```

```python
data_obs = np.array([[1], [1]])
student_params = np.array([0, 1, 1]*2)
student_em = emission_model.Student_Emission(2, data_obs)
```

```python
n_ar = np.array([0.5,1, 2, 4, 8])

x = np.linspace(-10, 10, 1000)
for n in n_ar:
    y = student_em.Student(x, 0, 1, n)
    plt.plot(x, y)
```

```python
 # sample the student distribution with accept_reject and check the variance vs. n (Diverges for n<= 2!!)
n_ar = [3, 4, 5, 6, 7, 8, 9, 10]
x_list = []
for n in n_ar:
    N = 0
    x_n = []
    while(N<10000):
        
        xj = np.random.rand() * 60 - 30
        yj = np.random.rand() * student_em.Student(0, 0, 1, n)
        
        if yj < student_em.Student(xj, 0, 1, n):

           x_n.append(xj)
           N+=1

    x_list.append(np.array(x_n))
```

```python
var = []
for i,n in enumerate(n_ar):

    var.append(np.var(x_list[i]))

plt.plot(n_ar, var, marker='o', linestyle='')
x= np.linspace(2.7, 10, 100)
plt.plot(x, (x)/(x-2))
```
